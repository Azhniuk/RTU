```python
import numpy as np
import matplotlib.pyplot as plt
import math as math

img = np.ones((1000,1000,3))

# The coordinates of the points of the object
point1 = (300,100)
point2 = (100,100)
point3 = (100,300)
point4 = (300,300)


#translate
def MoveObject(move_x, move_y):
    x1 = point1[0] + move_x
    y1 = point1[1] + move_y
    x2 = point2[0] + move_x
    y2 = point2[1] + move_y
    x3 = point3[0] + move_x
    y3 = point3[1] + move_y
    x4 = point4[0] + move_x
    y4 = point4[1] + move_y
    
    xArr = [x1,x2,x3,x4,x1]
    yArr = [y1,y2,y3,y4,y1]
    plt.plot(xArr,yArr)
    return
 
    
def ScaleObject(scale_x, scale_y):
    x1 = point1[0] * scale_x
    y1 = point1[1] * scale_y
    x2 = point2[0] * scale_x
    y2 = point2[1] * scale_y
    x3 = point3[0] * scale_x
    y3 = point3[1] * scale_y
    x4 = point4[0] * scale_x
    y4 = point4[1] * scale_x
    
    xArr = [x1,x2,x3,x4,x1]
    yArr = [y1,y2,y3,y4,y1]
    plt.plot(xArr,yArr)
    return

def RotateObject(angle):
    x1 = point1[0] * math.cos(angle) - point1[1]*math.sin(angle)
    y1 = point1[0] * math.sin(angle) + point1[1]*math.cos(angle)
    
    x2 = point2[0] * math.cos(angle) - point2[1]*math.sin(angle)
    y2 = point2[0] * math.sin(angle) + point2[1]*math.cos(angle)
    
    x3 = point3[0] * math.cos(angle) - point3[1]*math.sin(angle)
    y3 = point3[0] * math.sin(angle) + point3[1]*math.cos(angle)
    
    x4 = point4[0] * math.cos(angle) - point4[1]*math.sin(angle)
    y4 = point4[0] * math.sin(angle) + point4[1]*math.cos(angle)
    
    xArr = [x1,x2,x3,x4,x1]
    yArr = [y1,y2,y3,y4,y1]
    plt.plot(xArr,yArr)
    return
    

plt.figure(figsize=(10,8), dpi = 100, facecolor='w')
plt.imshow(img)
    
MoveObject(200,200)
ScaleObject(2, 2)

RotateObject(0.9599310886)



```
