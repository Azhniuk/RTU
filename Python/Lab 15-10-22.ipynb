Use the Bresenham algorithm to draw the last 3 digits of your student ID number using multiple straight lines 
for example, if your student ID number is 200RDB153, then you have to draw the digits 153).


```python
import matplotlib.pyplot as plt
import math
plt.title("Bresenham Algorithm")
plt.xlabel("x axis")
plt.ylabel("y axis")





def lineCreation (xBeg, yBeg, xEnd, yEnd):
    xCoordinates = [xBeg]
    yCoordinates = [yBeg]
    dx = abs(xEnd - xBeg)
    dy = abs(yEnd - yBeg)
    x = xBeg
    y = yBeg
    
    
    #Main variables
    if xBeg < xEnd :
        xs = 1
    else:
        xs = -1

    if yBeg < yEnd :
        ys = 1
    else:
        ys = -1
        
        
    #Steep or not
    if dx > dy:
        p = 2*dy - dx
        pn = p
        while x!=xEnd:
            x = x + xs
            if p > 0:
                y = y + ys
                p=p+2*dy-2*dx
            else:
                p=p+2*dy
            xCoordinates.append(x)
            yCoordinates.append(y)       
        
    else:
        p = 2*dx - dy
        pn = p
        while y!=yEnd:
            y = y + ys
            if p > 0:
                x = x + xs
                p=p+2*dx-2*dy
            else:
                p=p+2*dx
            xCoordinates.append(x)
            yCoordinates.append(y)    
            
            
            
    plt.plot(xCoordinates, yCoordinates)
                  
        
    return 




         
                
#2
lineCreation (0, 400, 100, 600)
lineCreation (100, 600, 300, 600)
lineCreation (0, 0, 300, 600)
lineCreation (0, 0, 300, 0)

#0
lineCreation (350, 600, 550, 600)
lineCreation (550, 600, 550, 0)
lineCreation (350, 0, 350, 600)
lineCreation (350, 0, 550, 0)

#7
lineCreation (600, 600, 900, 600)
lineCreation (900, 600, 600, 0)
lineCreation (600, 300, 900, 300)

plt.show()
  
```
