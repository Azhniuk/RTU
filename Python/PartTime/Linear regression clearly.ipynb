```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 1.5, 2.5, 4, 4.5, 5.7, 6]

plt.scatter(x = x, y = y)
plt.show()
```

```python
a = np.polyfit(x, y, 1)
#[slope, y-intersection]
p = np.poly1d(a)
#функція виду y = kx + b
print(a)
print(p)
```

```python
plt.scatter(x = x, y = y)
plt.plot(x, p(x))
#plot(x, f(x)) - х залишається, у - за знайденою формулою
plt.show()
```
